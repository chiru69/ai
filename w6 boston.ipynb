WEEK 6 BOSTON DATASET (LINEAR REGRESSION)

```python
from boston import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
```

```python
boston_datasets = load_boston()
```

```python
boston = pd.DataFrame(boston_datasets.data , columns = boston_datasets.feature_names)
boston
```

```python
boston.head()
```

```python
boston.describe()
```

```python
boston.isnull().sum()
```

```python
boston['PRICE']=boston_datasets.target
```

```python
boston
```

```python
cm = boston.corr().round(2)
sns.heatmap(cm, annot = True)
```

```python
plt.figure(figsize = (20,8))
feature = ['LSTAT','RM']
target = boston.PRICE
for i , col in enumerate(feature):
    plt.subplot(1,len(feature),i+1)
    plt.scatter(boston[col],target)
```

```python
x = pd.DataFrame(np.c_[boston['LSTAT'],boston['RM']],columns = ['LSTAT','RM'])
y = boston.PRICE
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
ytp = model.predict(x_train)
rmse = np.sqrt(mean_squared_error(y_train,ytp))
r2 = r2_score(y_train,ytp)
print('Training')
print(rmse)
print(r2)
```

```python
ytp = model.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test,ytp))
r2 = r2_score(y_test,ytp)
print('Testing')
print(rmse)
print(r2)
```
